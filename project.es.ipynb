#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
from scipy.stats import zscore
sub_ds.loc[:, "price_zscore"] = sub_ds.groupby("level5")["price"].transform(zscore).copy()
plt.figure(figsize=(10,6))
for poblacion in sub_ds["level5"].unique():
    subset = sub_ds[sub_ds["level5"] == poblacion]
    plt.hist(subset["price_zscore"], bins=20, alpha=0.6, label=poblacion)
plt.title("Histogramas normalizados de precios por población (z-score)")
plt.xlabel("Precio normalizado")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Al normalizar los precios por población, podemos ver cómo se comportan las casas dentro de cada zona en comparación con su propia media. En general, la mayoría de las viviendas están cerca del promedio, pero hay algunas que destacan por ser mucho más caras.

Getafe llama la atención porque tiene como dos grupos de precios marcados, lo que podría significar que hay zonas más exclusivas dentro del mismo municipio. En cambio, Fuenlabrada, Alcorcón y Leganés tienen precios más concentrados, lo que indica que hay menos diferencia entre las casas de esas zonas.

En resumen, aunque todas las poblaciones tienen casas más caras de lo normal, Getafe parece tener más variedad, mientras que las demás son más “uniformes” en sus precios.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
import matplotlib.pyplot as plt
poblaciones = sub_ds["level5"].unique()
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 filas, 2 columnas
axs = axs.flatten()
for i, poblacion in enumerate(poblaciones):
    datos = sub_ds[sub_ds["level5"] == poblacion]
    axs[i].scatter(datos["surface"], datos["price"], alpha=0.6)
    axs[i].set_title(poblacion)
    axs[i].set_xlabel("Superficie (m²)")
    axs[i].set_ylabel("Precio (€)")
plt.tight_layout()
plt.show()
```
